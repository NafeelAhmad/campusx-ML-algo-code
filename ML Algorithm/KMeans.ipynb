```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r'C:\Users\hp\Downloads\Wholesale customers data.csv')
df.head(5)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)

new_df = pd.DataFrame(data_scaled)
new_df.describe()

```

```python
kmeans = KMeans(n_clusters=2,init='k-means++')
kmeans.fit(new_df)
```

```python
kmeans.inertia_
```

```python
SSE = []

for n_clusts in range(1,20):
    model_fit = KMeans(n_clusters=n_clusts,init='k-means++')
    model_fit.fit(new_df)
    SSE.append(model_fit.inertia_)

print(SSE)
```

```python
elb_data = pd.DataFrame({'Clusters':range(1,20),'SSE':SSE})

plt.plot(elb_data['Clusters'],elb_data['SSE'],marker='o')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
plt.title('Optimal Value of K')
```

```python
# k means using 5 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 7, init='k-means++')
kmeans.fit(new_df)
pred = kmeans.predict(new_df)
```

```python
final_df = pd.DataFrame(new_df)
final_df['Cluster'] = pred
final_df.Cluster.value_counts()
```
