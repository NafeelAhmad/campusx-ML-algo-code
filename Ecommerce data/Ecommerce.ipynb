```python
# data type change required
# Removal of Null Values
# Outliers Inspection
# Transformation Required : Calculated Column(Total Amount Spent, Year, Month, Date Seperation) 
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0,subset="CustomerID")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df[df['Quantity'] >= 0]
```

```python
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df.info()
```

```python
df['CustomerID'] = df['CustomerID'].astype(str)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'],errors='coerce')
```

```python
df.head()
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
df['DayName'] = df['InvoiceDate'].dt.day_name()
df['Date'] = df['InvoiceDate'].dt.date
```

```python
df.head()
```

```python
df['TotalSpent'] = df['UnitPrice']*df['Quantity']
```

```python
df.head()
```

```python
# Data Cleaning, Data Formatting and Data Transformation have done till this stage.
```

```python
""" Now we will try to findout the answer of the objectives 
and figure out the hidden pattern from the data.  """
```

```python
country_wise_totalspent = df.groupby('Country')['TotalSpent'].sum().sort_values(ascending=False).head(10)
country_wise_totalspent
```

```python
plt.figure(figsize=(15,6))
plt.bar(country_wise_totalspent.index,country_wise_totalspent.values)
plt.title('Total Spent By Country-wise')
plt.xlabel('Country')
plt.ylabel('Total Spent')
#plt.bar_label(bars, padding=3)
#plt.margins(y=0.1)
#plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Set the figure size first
plt.figure(figsize=(15, 6))

# 2. Create the plot
bars = plt.bar(country_wise_totalspent.index, country_wise_totalspent.values)

# 3. Add labels and title
plt.title('Total Spent By Country')
plt.xlabel('Country')
plt.ylabel('Total Spent')

# 4. Add bar labels and adjust margins
plt.bar_label(bars, padding=3, fmt='{:,.0f}') # Added formatting for cleaner numbers
plt.margins(y=0.1)

# 5. Rotate x-axis labels if you have many countries
plt.xticks(rotation=45) 

plt.show()
```
