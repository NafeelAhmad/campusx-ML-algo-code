```python
# Singleton Pattern

class ControlTower:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            print('Initializing Control Tower')
        return cls._instance
    def __init__(self):
        print('Learning Singleton Pattern')
    
Tower1 = ControlTower()
Tower2 = ControlTower()

print(Tower1 is Tower2)
print(id(Tower1))
print(id(Tower2))
```

```python
# Observer Pattern

from typing import List
class Stock:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.observers: List = []  # List to keep track of subscribers

    def add_observer(self, observer):
        self.observers.append(observer)

    def remove_observer(self, observer):
        self.observers.remove(observer)

    def notify_observers(self):
        for observer in self.observers:
            # All observers must follow a common interface (the update method)
            observer.update(self)

    def set_price(self, price):
        self.price = price
        # Automatically trigger notification when state changes
        self.notify_observers()

# Observer Interface/Base Class
class Observer:
    def update(self, stock):
        raise NotImplementedError("This Method should overridden by subclasses")

# Concrete Observers
class Dashboard(Observer):
    def update(self, stock):
        print(f"Dashboard: {stock.name} price updated to {stock.price}")

class EmailAlerts(Observer):
    def update(self, stock):
        print(f"Email: {stock.name} price alert! New price: {stock.price}")

class SMSAlerts(Observer):
    def update(self, stock):
        print(f"SMS: {stock.name} is now {stock.price}")

# Usage [00:27:00]
apple_stock = Stock("Apple", 150)

# Create observer instances
dashboard = Dashboard()
email = EmailAlerts()
sms = SMSAlerts()

# Subscribe observers to the stock
apple_stock.add_observer(dashboard)
apple_stock.add_observer(email)
apple_stock.add_observer(sms)

# Changing the price automatically notifies all subscribed observers
apple_stock.set_price(155)
```

```python
# Builder Pattern 



class House:
    def __init__(self, bedrooms, bathrooms, kitchen, garden, garage, pool, solar_panels, smart_home):
        self.bedrooms = bedrooms
        self.bathrooms = bathrooms
        self.kitchen = kitchen
        self.garden = garden
        self.garage = garage
        self.pool = pool
        self.solar_panels = solar_panels
        self.smart_home = smart_home

    def __str__(self):
        return f"House with {self.bedrooms} bedrooms, {self.bathrooms} bathrooms, Kitchen: {self.kitchen}, Garden: {self.garden}, Garage: {self.garage}, Pool: {self.pool}, Solar: {self.solar_panels}, Smart Home: {self.smart_home}"

class HouseBuilder:
    def __init__(self):
        # Default values
        self.bedrooms = 0
        self.bathrooms = 0
        self.kitchen = True  # Basic necessity
        self.garden = False
        self.garage = False
        self.pool = False
        self.solar_panels = False
        self.smart_home = False

    def set_bedrooms(self, count):
        self.bedrooms = count
        return self  # Return self to allow method chaining

    def set_bathrooms(self, count):
        self.bathrooms = count
        return self

    def add_garden(self):
        self.garden = True
        return self

    def add_garage(self):
        self.garage = True
        return self

    def add_pool(self):
        self.pool = True
        return self

    def add_solar_panels(self):
        self.solar_panels = True
        return self

    def build(self):
        # The final step that returns the actual House object
        return House(
            self.bedrooms, self.bathrooms, self.kitchen, self.garden, 
            self.garage, self.pool, self.solar_panels, self.smart_home
        )

# Usage [00:40:27]
builder = HouseBuilder()
custom_house = (builder.set_bedrooms(4)
                       .set_bathrooms(3)
                       .add_garage()
                       .add_pool()
                       .add_solar_panels()
                       .build())

print(custom_house)
```

```python
# Strategy Pattern


# Strategy Interface
class PaymentStrategy:
    def pay(self, amount):
        pass

# Concrete Strategies
class CreditCardPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Processing ${amount} payment using Credit Card")

class PayPalPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Processing ${amount} payment using PayPal")

class CryptoPayment(PaymentStrategy):
    def pay(self, amount):
        print(f"Processing ${amount} payment using Cryptocurrency")

# Context Class [00:48:07]
class PaymentProcessor:
    def __init__(self, strategy):
        self.strategy = strategy

    def set_strategy(self, strategy):
        # Allows switching algorithms at runtime
        self.strategy = strategy

    def process_payment(self, amount):
        self.strategy.pay(amount)

# Usage [00:48:35]
# Start with Credit Card
processor = PaymentProcessor(CreditCardPayment())
processor.process_payment(100)
processor.strate

# Switch to PayPal at runtime
# processor.set_strategy(PayPalPayment())
# processor.process_payment(200)

# # Switch to Crypto
# processor.set_strategy(CryptoPayment())
# processor.process_payment(300)
```
