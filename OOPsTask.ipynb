### Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
class Circle:

    def __init__(self,radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
    
    def parameter(self):
        return 2 * 3.14 * self.radius
```

```python
c1 = Circle(4)
c1.area()
c1.parameter()
```

### Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.

```python
class Person:

    def __init__(self,name,country,dob_year,dob_month,dob_day):
        self.name = name
        self.country = country
        self.dob_year = dob_year
        self.dob_month = dob_month
        self.dob_day = dob_day
    

    from datetime import date
    def age(self):
        today = date.today()
        age = today.year - self.dob_year
        age -= (self.dob_year , self.dob_month) < (today.year,today.month)
        return age
```

```python
p1.age()
```

### Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
class Calculator:

    def __init__(self,n1,n2):
        self.n1 = n1
        self.n2 = n2

    def addition(self):
        return self.n1 + self.n2
    
    def subtraction(self):
        return self.n1 - self.n2

    def multiplication(self):
        return self.n1 * self.n2

    def division(self):
        if self.n2 != 0:
            return self.n1 / self.n2
        else:
            return 0
```

### Write a Program with class Employee that keeps a track of the number of employees in an organizatioin and also stores their name, designation, and salary details

```python
class Employee():

    counter = 0
    def __init__(self,name,designation,salary):
        self.name = name
        self.designation = designation
        self.salary = salary

        Employee.counter += 1

    def employee_count(self):
        return f'Total Employees in the organisation are {Employee.counter}'

    def employee_details(self):
        return f'Name: {self.name}, Designation: {self.designation}, Salary: {self.salary} '

```

### Write a program that has a class Person storing name and date of birth of a person. The Program should subtract the DOB from today's date to findout whether a person is eligible to vote or not.

```python
class Person:

    def __init__(self,name,dob_year,dob_month,dob_day):
        self.name = name
        self.__dob_year = dob_year
        self.__dob_month = dob_month
        self.__dob_day = dob_day

    def is_eligible(self):
        from datetime import date
        today = date.today()
        age = today.year - self.__dob_year

        age -= (today.month,today.day) > (self.__dob_month,self.__dob_day)
        if age >= 18:
            return 'Eligible to Vote'
        else: 
            return 'Not Eligible to Vote'

```

```python
p1 = Person('Nafeel',2002,5,26)
```

```python
p1.is_eligible()
```
