```python
n = int(input())
s = set(map(int, input().split()))

N = int(input())

for i in range(1,N+1):
    b = list(map(str, input().split()))
    
    
    if len(b) == 1 and b == "pop":
        s.pop()
    elif b[0] == "remove":
        s.remove(int(b[1]))
    elif b[0] == "discard":
        s.discard(int(b[1]))

sum(s)
```
