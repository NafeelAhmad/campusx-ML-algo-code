### The "E-Commerce" Challenge: Inventory & Sales Tracker
The Scenario:
You are managing an e-commerce inventory system. Each product is a key, and its details (price, quantity) are stored in a nested dictionary.

```python
inventory = {
    "Laptop": {"price": 1000, "stock": 5},
    "Mouse": {"price": 25, "stock": 50},
    "Keyboard": {"price": 75, "stock": 20}
}
```

Retrieval: Retrieve the price of a "Laptop". If the item does not exist, use .get() to return a message like "Item not found" instead of crashing.

Updating: A shipment of 10 new Keyboards arrived. Update the "stock" value for "Keyboard" by adding 10 to the current value.

Adding New Items: Add a new product, "Monitor", with a price of 300 and stock of 10.

Dictionary Comprehension: Create a new dictionary called discounted_items that contains the items from inventory but with their prices reduced by 10% (use a comprehension).

Iteration: Write a for loop to print each product name and its total value (price * stock) in the inventory.

Keys/Values/Items: Extract all product names (keys) into a list, and all product details (values) into a list.

Frequency Counter (The Classic): Given a list of recent sales sales_list = ["Laptop", "Mouse", "Laptop", "Laptop", "Keyboard"], use a dictionary to count how many of each item was sold.

```python
{i:{**j,'price':j['price']*0.9} for i,j in inventory.items()}
```

```python
# Create a new dict with 10% off prices
discounted_items = {
    product: {**details, "price": details["price"] * 0.9} 
    for product, details in inventory.items()
}

print(discounted_items)
# Output: {'Laptop': {'price': 900.0, 'stock': 5}, ...}
```

```python
for i,j in inventory.items():
    print(f'{i}: ${j['price']*j['stock']}')
```

```python
product_name = [i for i,j in inventory.items()]
print(product_name)
product_list = [j for i,j in inventory.items()]
print(product_list)
```

```python
recent_sales_list = ['Laptop','Mouse','Laptop','Laptop','Keyboard']
sales_count = {}

for item in recent_sales_list:
    if item in sales_count:
        sales_count[item] += 1
    else:
        sales_count[item] = 1

print(sales_count)
    
```
