## The "Data Integrity" Challenge: Weather Records

The Scenario:
You are developing a weather tracking system. Since historical data should not be changed, you must use tuples to store records. Each record is a tuple: (City, Temperature, Humidity).

```python
# A collection of records
weather_data = [
    ("Bengaluru", 28, 65),
    ("London", 15, 80),
    ("New York", 22, 55),
    ("Tokyo", 25, 70)
]
```

The Tasks
Packing & Unpacking: Create a single tuple named new_record representing ("Dubai", 40, 20). Then, unpack this tuple into three separate variables: city, temp, and humidity.

Immutability Observation: Attempt to change the temperature of "Bengaluru" from 28 to 30 by accessing the index. Write a comment explaining why this operation raises a TypeError.

Conversion & Re-creation: Since you cannot modify a tuple, create a new version of the Bengaluru record with the updated temperature (30) by converting the tuple into a list, modifying the value, and converting it back to a tuple.

Tuple Slicing: Create a new tuple containing only the (Temperature, Humidity) portions of all entries in weather_data using a list comprehension.

Concatenation: Combine weather_data with a new tuple extra_data = (("Paris", 18, 75),) to create a single master collection.

Search & Count: Determine how many cities in your weather_data have a humidity level greater than 60 by iterating through the collection.

Zipping: Use the zip() function to pair a list of dates = ["2026-04-26", "2026-04-27"] with the first two tuples in your weather_data to create a list of combined tuples.
