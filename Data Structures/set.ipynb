### The "Master Data" Challenge: Employee Access Control
The Scenario:
You are working on a security system for an office building. You have two lists of employees: one list of people who swiped into the Office and another list of people who swiped into the Server Room.

```python
office_log = {"Alice", "Bob", "Charlie", "David", "Eve", "Frank"}
server_room_log = {"Charlie", "Eve", "Grace", "Heidi"}
```

The Tasks
Unique Tracking: A new employee, "Ivan", swiped into the office. Add him to the office_log.

Intersection: Find the list of employees who were in both the office and the server room.

Difference: Find the employees who were in the office but did not enter the server room.

Union: Create a comprehensive list of all unique individuals who accessed any part of the building.

Symmetric Difference: Identify employees who were in either the office or the server room, but not both (who might have bypassed a security protocol).

Membership Testing: Write a quick check to see if "David" is present in the server_room_log.

Data Cleaning: You receive a new list of employees: ["Alice", "Alice", "Bob", "Grace", "Grace"]. Convert this to a set to automatically remove the duplicates.

```python
#Task1
office_log.add('Ivan')
```

```python
office_log
```

```python
#task2 

common_employees = office_log.intersection(server_room_log)
print(common_employees)
```

```python
#task3

employees_accessed_any_part_of_building = office_log.union(server_room_log)
print(employees_accessed_any_part_of_building)
```

```python
#task4

employees_in_office_not_in_server_room = office_log.difference(server_room_log)
print(employees_in_office_not_in_server_room)
```

```python
#task5

either_in_office_or_sever_room = office_log.symmetric_difference(server_room_log)
print(either_in_office_or_sever_room)
```

```python
#task6

if 'David' in server_room_log:
    print('Present')
else: 
    print('Not Present')
```

```python
#task7

Data_Cleaning = ["Alice","Alice", "Bob","Grace","Grace"]
cleaned_data = set(Data_Cleaning)
print(cleaned_data)
```
