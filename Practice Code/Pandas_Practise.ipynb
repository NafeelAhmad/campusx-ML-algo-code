```python
import pandas as pd

df = pd.read_csv(r'G:\campusx ML algo code\Ecommerce data\data.csv', encoding='ISO-8859-1')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns='StockCode')
```

```python
df = df.dropna(subset='CustomerID')
```

```python
df.isnull().sum()
```

```python
df['Quantity'].describe()
```

```python
df = df[df['Quantity']>0]
```

```python
df.dtypes
```

```python
df['CustomerID'] = df['CustomerID'].astype('str')
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.dtypes
```

```python
df.describe(include='all')
```

```python
df['Total_Amount'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Date'] = df['InvoiceDate'].dt.date
df['Day_Name'] = df['InvoiceDate'].dt.day_name()
```

```python
res1 = df.groupby(['Country'])['Total_Amount'].sum().reset_index()
res1.sort_values(by='Total_Amount',ascending=True)
```
