```python
class Rectangle:

    def __init__(self,width,length):
        self.width = width
        self.length = length

    def area(self):
        return self.width * self.length

    def perimeter(self):
        return 2*(self.width + self.length)

Rec_class = Rectangle(10,4)
print("Area =", Rec_class.area() , ", Perimeter =",Rec_class.perimeter())
        
```

```python
class Student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def average(self):
        return round(sum(self.marks)/len(self.marks),1)

st_class = Student("Alice", [85, 90, 78, 92, 88])
print(f'{st_class.name} Average Grade : {st_class.average()}')
```

```python
class BankAccount:

    def __init__(self):
        self.balance = 1000
        

    def deposit(self,amount):
        self.balance += amount
        return "Balance After Deposit: ",self.balance

    def withdraw(self,wdamount):
        if self.balance >= wdamount:
            self.balance -= wdamount
            return "Balance After Withdrawl: ",self.balance
        else:
            return "Insufficient Balance, Current Balance",self.balance


BA_class = BankAccount()
print(BA_class.deposit(500))
print(BA_class.withdraw(200))
print(BA_class.withdraw(2000))
#print(BA_class.withdraw(2000))
```

```python
class Light:

    def __init__(self):
        self.state = None

    def turn_on(self):
        self.state = 'ON'
        print('Light is ON')

    def turn_off(self):
        self.state = 'OFF'
        print('Light is OFF')
    
    def status(self):
        print("Current Status:",self.state)

lg_class = Light()
lg_class.turn_on()
lg_class.status()
lg_class.turn_off()
lg_class.status()
```

```python
class Employee:

    def __init__(self,name):
        self.name = name
        

class FullTimeEmployee(Employee):

    def __init__(self,name,salary):
        super().__init__(name)
        self.salary = salary

    def pay(self):
        print(f"{self.name}'s monthly pay: {self.salary}")

class PartTimeEmployee(Employee):

    def __init__(self,name,perday_salary,days):
        super().__init__(name)
        self.perday_salary = perday_salary
        self.days = days

    def pay(self):
        print(f"{self.name}'s monthly pay: {self.perday_salary*self.days}")


e1 = FullTimeEmployee("Alice", 60000)
e2 = PartTimeEmployee("Bob", 500, 20)

e1.pay()
e2.pay()
```

```python
class Shape:

    def area(self):
        pass

class Circle(Shape):

    def __init__(self,radius):
        self.radius = radius

    def area(self):
        print(3.14*self.radius*self.radius)

class Square(Shape):

    def __init__(self,side):
        self.side = side
        
    def area(self):
        print(self.side*self.side)

class Triangle(Shape):

    def __init__(self,base,height):
        self.base = base
        self.height = height

    def area(self):
        print(0.5*self.base*self.height)

diff_shapes = [Circle(7),Square(4),Triangle(6, 8)]

for sh in diff_shapes:
    sh.area()
# s1.area()
# s2 = Square(4)
# s2.area()
# s3 = Triangle(6, 8)
# s3.area()
```

```python
class Order:

    def __init__(self,id,Amount):
        self.id = id
        self.Amount = Amount
        
    def get_total(self):
        print(self.Amount)

class DiscountedOrder(Order):

    def __init__(self,id,Amount):
        super().__init__(id,Amount) 
        
        
    def get_total(self):
        print("Order ID:",self.id)
        print("Original Total:",self.Amount)
        print("Discounted Total:",self.Amount*0.9)

ord = DiscountedOrder("ORD001", 1200)
ord.get_total()
```
