```python
class Rectangle:

    def __init__(self,width,length):
        self.width = width
        self.length = length

    def area(self):
        return self.width * self.length

    def perimeter(self):
        return 2*(self.width + self.length)

Rec_class = Rectangle(10,4)
print("Area =", Rec_class.area() , ", Perimeter =",Rec_class.perimeter())
        
```

```python
class Student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def average(self):
        return round(sum(self.marks)/len(self.marks),1)

st_class = Student("Alice", [85, 90, 78, 92, 88])
print(f'{st_class.name} Average Grade : {st_class.average()}')
```

```python
class BankAccount:

    def __init__(self):
        self.balance = 1000
        

    def deposit(self,amount):
        self.balance += amount
        return "Balance After Deposit: ",self.balance

    def withdraw(self,wdamount):
        if self.balance >= wdamount:
            self.balance -= wdamount
            return "Balance After Withdrawl: ",self.balance
        else:
            return "Insufficient Balance, Current Balance",self.balance


BA_class = BankAccount()
print(BA_class.deposit(500))
print(BA_class.withdraw(200))
print(BA_class.withdraw(2000))
#print(BA_class.withdraw(2000))
```

```python
class Light:

    def __init__(self):
        self.state = None

    def turn_on(self):
        self.state = 'ON'
        print('Light is ON')

    def turn_off(self):
        self.state = 'OFF'
        print('Light is OFF')
    
    def status(self):
        print("Current Status:",self.state)

lg_class = Light()
lg_class.turn_on()
lg_class.status()
lg_class.turn_off()
lg_class.status()
```

```python
class User:

    def __init__(self,username,password):
        self.username = username
        self.__password = password

    def check_password(self,input_password):
        if self.__password == input_password:
            print(True)
        else:
            print(False)

u1 = User("alice", "secure123")
u1.check_password("secure123")

```

```python
class Notebook:

    def __init__(self):
        self.notes = []

    def add_note(self,note):
        self.notes.append(note)
        

    def show_notes(self):
        self.notes
        for i,j in enumerate(self.notes):
            print(f'{i+1}. {j}')
        

    
u1 = Notebook()
u1.add_note("Buy groceries")
u1.add_note("Read a book")
u1.add_note("Call the doctor")

u1.show_notes()

```
